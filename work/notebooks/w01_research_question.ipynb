## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

```python
What this work can claim:
This clustering work can say that certain groups of pages observed similar patterns in the data — for example, pages with high impressions and stable engagement clustering together, versus pages with low impressions and declining trends clustering separately. These clusters are meant to be decision-support: they suggest which pages a reviewer should look at first, not a final answer. Any conclusion drawn from this analysis should be treated as directional — a pattern the data suggests, not a guaranteed outcome.
What this work cannot claim:
This work cannot prove that any action (like refreshing a page) will cause traffic to recover — that would require a controlled experiment, which this data does not provide. It also cannot claim to understand or predict Google's search ranking algorithm; the data only reflects observed outcomes, not the reasons behind them. Since the dataset contains only numeric and categorical metrics (no actual article text), these clusters should never be called "semantic clustering" — they are based on behavior and metadata, not meaning. Finally, these clusters are not fixed truths; if the features or number of clusters change, the groupings may look different, so they should be treated as a useful lens rather than a permanent classification.
```
